```python
import matplotlib.pyplot as plt
```

## Plotting Scatter Points

```python
x=[1,2,3]
y=[2,4,6]
plt.scatter(x,y)
plt.show()
```

## Plotting Graph

```python
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y)
plt.scatter(x,y)
```

```python
x1=[1,2,3]
y1=[2,4,6]

x2=[10,20,30]
y2=[2,4,6]

plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

## Setting Color & Markers

```python
x=[1,2,3]
y=[2,4,6]

plt.plot(x,y,'ro')
plt.show()
```

## Plotting Functions

```python
import numpy as np
x=np.array([1,2,3,4])
x=np.arange(0,5,0.1) # x=[0,0.1,0.2,....,4.9]
y=x**3
plt.plot(x,y,'b--')
plt.show()
```

## Passing Single arg

```python
a=[4,8,10,12,15]
plt.plot(a,'bp')
plt.show()
# By default it takes y=a & x=[0,1,2,3,4,5] as there are 5 nos
```
