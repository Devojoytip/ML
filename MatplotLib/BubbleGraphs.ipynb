```python
import matplotlib.pyplot as plt
```

```python
year=[2001,2002,2003,2004,2005,2006,2007,2008]
salary=[50,100,200,600,1000,1200,1500,2000]
population=[100,200,300,350,380,500,100,55]

plt.title('Bubble Graphs')
plt.xlabel('Years')
plt.ylabel('Salary')
plt.scatter(year,salary,s=population,)
```

## Customising

```python
# setting color
plt.scatter(year,salary,s=population,c='r')
```

```python
# setting diff color for diff bubbles
import numpy as np
c=np.arange(len(year))
plt.scatter(year,salary,s=population,c=c)
```

```python
# setting opacity
plt.scatter(year,salary,s=population,c='r',alpha=0.1)
```

```python

# setting edgecolor
plt.scatter(year,salary,s=population,c='r',edgecolor='g')
```

```python
# setting marker
# refer : https://matplotlib.org/stable/api/markers_api.html
plt.scatter(year,salary,s=population,c='r',edgecolor='g',marker='s')
```
