```python
import matplotlib.pyplot as plt
```

```python
sizes=[25,15,10,50]
plt.pie(sizes)
```

## Decoration

```python
labels=['A','B','C','D']
plt.title='Pie Graph'
colors=['red','blue','orange','green']
# if u want certain pie to separate out
explode=[0.3,0,0.2,0]
# autopct to calculate % value of each piece upto desired decimal points
# pie graph is by default in counter clockwise dir which can be set False
# startangle - angle at which 1st entry(here A) starts , default - 0
plt.pie(sizes,colors=colors,explode=explode,labels=labels,autopct='%.2f',counterclock=False,startangle=90)
plt.show()
```
