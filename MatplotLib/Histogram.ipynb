```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=[1,2,2,3,3,4,1,1,5,4,1,2]
plt.hist(a)
```

```python
a=[1,2,1,3,8,5,6,7,5,10,10,12,10,22,20,20,21,15,15,19,18,16,10,22,7,9]
xt=np.arange(23)
print(xt)
plt.hist(a,edgecolor='red')
plt.xticks(xt)
plt.show()
```

## Bins

```python
plt.hist(a,edgecolor='blue',bins=max(a)-min(a))
plt.xticks(xt)
plt.show()
```
