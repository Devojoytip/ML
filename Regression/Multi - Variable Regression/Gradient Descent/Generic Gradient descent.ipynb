```python
import numpy as np
```

```python
from sklearn import datasets
data = datasets.fetch_california_housing()
X = data.data
Y = data.target
```

```python
def cost(m, X, Y):
    total_cost = 0
    p = len(X)
    for i in range(p):
        total_cost += (2/p)*(Y[i]-sum(m*X[i]))
    return total_cost
```

```python
def step_gradient(X, Y, learning_rate, m):
    n = len(X[0])
    m_slope = np.array([0 for i in range(n)])
    p = len(X)

    for i in range(p):
        x = X[i]
        y = Y[i]
        for j in range(n):
            m_slope[j] += (-2/p)*(y-((m*x).sum()))*x[j]  # calc mj 
        new_m = m-learning_rate*m_slope
        return new_m
```

```python
def gd(X, Y, learning_rate, num_iterations):
    m = np.array([0 for i in range(len(X[0]))])  # slope array

    for i in range(num_iterations):
        m = step_gradient(X, Y, learning_rate, m)
        print(i, " Cost: ", cost(m, X, Y))
    return m
```

```python
def run():
    X = data.data
    Y = data.target
    X = np.append(X, np.ones(len(X), dtype='int').reshape(-1, 1), axis=1)
    iterations = 500
    learning_rate = 0.00001
    m = gd(X, Y, learning_rate, iterations)
    print(m)
```

```python
run()
```
