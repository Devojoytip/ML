```python
import numpy as np
```

```python
data=np.loadtxt("data.csv",delimiter=",")
```

```python
x=data[:,0].reshape(-1,1)
y=data[:,1]
```

```python
x.shape
```

```python
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(x,y)
```

```python
from sklearn.linear_model import LinearRegression
algo=LinearRegression()
algo.fit(X_train,Y_train)
```

```python
algo.coef_
```

```python
algo.intercept_
```

```python
import matplotlib.pyplot as plt
m=algo.coef_[0]
c=algo.intercept_

x_line=np.arange(30,70,0.1)
y_line=m*x_line+c
plt.plot(x_line,y_line,"red")
train_ld=X_train.reshape(75) # converting back to 1D array
plt.scatter(X_train,Y_train)
plt.show()
```

```python
import matplotlib.pyplot as plt
m=algo.coef_[0]
c=algo.intercept_

x_line=np.arange(30,70,0.1)
y_line=m*x_line+c
plt.plot(x_line,y_line,"red")
train_ld=X_test.reshape(25) # converting back to 1D array
plt.scatter(X_test,Y_test)
plt.show()
```

```python
train_score=algo.score(X_train,Y_train)
test_score=algo.score(X_test,Y_test)

train_score,test_score
```
