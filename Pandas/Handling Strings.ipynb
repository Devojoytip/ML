```python
import pandas as pd
```

```python
iris=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
df=iris.copy()
df.columns=['sl','sw','pl','pw','flower_type']
df['Gender']='Male'
df.describe()
df.index
```

```python
def fn(s):
    if(s=='Male'):
        return 1
    else:
        return 0
```

```python
# Applying a fn which takes input every entry of the col and returns some val
df["Sex"]=df["Gender"].apply(fn)
del df["Gender"]
df.head(149)
```
