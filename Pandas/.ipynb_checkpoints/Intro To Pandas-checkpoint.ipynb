## Pandas
Python Module for Data Manipulation

```python
import pandas as pd
```

```python
iris=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
print(type(iris)) # <class 'pandas.core.frame.DataFrame'>
iris
```

```python
df=iris # changes in df makes change in iris
df=iris.copy() # we make a copy of iris
```

```python
df.head()
df.head(10)
```

## Setting Col Names

```python
df.columns=['sl','sw','pl','pw','flower_type']
```

```python
print(df.shape) # (row,cols)
print(df.dtypes)
```

```python
df.describe()
```

```python
# dataframe.col_name for accessing a col
df.sl
```

```python
df.isnull()
df.isnull().sum() # lists no of null entries in each col
```

## Changing Values

```python
df.iloc[1:4,2:4]
```

```python
df.iloc[1:4,2:4]=1
```

## Deleting Rows Using .drop(label,inplace)

```python
# makes a copy of df without the row labelled n so no change in df
# but the indexing does not change after dropping rows
df.drop(0) 
df.drop(3,inplace=True)
df.head()
```

## Deleting Rows by index

```python
print(df.index[0])
print(df.index[3])
print(df.index[4])
```

```python
df.drop(df.index[3],inplace=True) # del row at idx = 3
df.drop(df.index[[2,3,4]],inplace=True) # del row at idx = 2,3,4
df.head()
```

## Passing boolean conditions

```python
df[df.sl>=5]
df[df.flower_type=='Iris-setosa'].describe()
```

## Adding Rows

```python
df.loc[0]=[1,2,3,4,'Rose']
```

```python
df.head()
df.tail()
```

```python
df.index
df.head()
```

## Resetting Indexes

```python
# df.reset_index() does n't make change to df & also adds a new col of prev indexes
# to drop extra col drop=True
df.reset_index(drop=True,inplace=True) 
df.head()
```

## Delete Cols

```python
df.drop('sl',axis=1,inplace=True)
df.head()
```
